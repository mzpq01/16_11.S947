```python
# Import some libraries
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# This allows plots to appear on the IPython notebook.
%matplotlib inline 
```

```python
# Use pandas to import the csv
df_edges = pd.read_csv('data/FoursqureLinksNetwork.csv', sep=",")
df_edges.head()
```

```python
df_nodes = pd.read_csv('data/FoursqureCheckinNodes.csv', sep=",")
df_nodes.head()
```

```python
df_nodes_latlon_from = df_nodes[['ID','LAT','LONG']]
df_nodes_latlon_from = df_nodes_latlon_from.rename(columns = {'ID':'FROM-ID','LAT':'FROM-LAT','LONG':'FROM-LON'})
df_nodes_indexed = df_nodes_latlon.set_index(['FROM-ID'])
df_nodes_indexed.head()

result_from = df_edges.join(df_nodes_indexed, on='FROM-ID')
result_from[:100]
```

```python
df_nodes_latlon_to = df_nodes[['ID','LAT','LONG']]
df_nodes_latlon_to = df_nodes_latlon_to.rename(columns = {'ID':'TO-ID','LAT':'TO-LAT','LONG':'TO-LON'})
df_nodes_latlon_to[:10]
df_nodes_indexed_2 = df_nodes_latlon_to.set_index(['TO-ID'])
df_nodes_indexed_2.head()

result_to = result_from.join(df_nodes_indexed_2, on='TO-ID')
result_to[:200]
```

```python
df_nodes_idlist = df_nodes['ID'].tolist()
df_nodes_idlist[:5]
print len(df_nodes['ID'])
print len(df_nodes_idlist)
print df_nodes['ID'][0]
node_attrs = df_nodes[['NAME','CATEGORY','CHECK-IN','LAT','LONG']]
```

```python
RG = nx.Graph()
node_attrs = df_nodes[['NAME','CATEGORY','CHECK-IN','LAT','LONG']]
for i in range(len(df_nodes['ID'])):
    RG.add_node(df_nodes['ID'][i], nodes_attr=node_attrs.iloc[[i]])
degrees = nx.degree(RG)    
degrees
all_nodes = RG.nodes()
nodes_num = RG.number_of_nodes()
print all_nodes
print nodes_num
```

```python
keys = ['DISTANCE','FROM-LAT','FROM-LON','TO-LAT','TO-LON']
for row in result_to.itertuples(index = False):
    values = row[2:]
    edge_attributes = dict(zip(keys, values))
    node_id = (row[0],row[1])
    RG.add_edge(*node_id, attr_dict=edge_attributes)
print RG.edges()
print RG.number_of_edges()
```
